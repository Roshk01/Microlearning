```python
def remove_space_from_column_names(df):
    new_column_names = [name.replace(" ", "") for name in df.columns]  # Replace space with underscore
    return df.rename(columns=dict(zip(df.columns, new_column_names)))  # Rename columns

# Remove spaces from column names
data2 = remove_space_from_column_names(data2.copy())  # Operate on a copy to avoid modifying original data
```

```python
def update_keywords(row):
    if not pd.isna(row['website']):
    # Add 'web' to 'keywords' with a space if existing keywords are present
        row['keyword'] = row['keyword'] + ', website' if not pd.isna(row['keyword']) else ""
    if not pd.isna(row['reference']):
    # Add 'reference' to 'keywords' with a space if existing keywords are present
        row['keyword'] = row['keyword'] + ', reference' if not pd.isna(row['keyword']) else ''

    if not pd.isna(row['wikipedia']):
    # Add 'wikipedia' to 'keywords' with a space if existing keywords are present
        row['keyword'] = row['keyword'] + ', wikipedia' if not pd.isna(row['keyword']) else ''
    if not pd.isna(row['documentation']):
    # Add 'document' to 'keywords' with a space if existing keywords are present
        row['keyword'] = row['keyword'] + ', document' if not pd.isna(row['keyword']) else ''

    return row
```

```python
df = df.apply(update_keywords, axis=1)
```

```python
df.head()
```

```python
def append_links(row):
    links = []

  # Add website if not null
    if not pd.isna(row['website']):
        links.append('website: ' + row['website'])

  # Add reference, documentation similarly (modify the column names as needed)
    if not pd.isna(row['reference']):
        links.append('reference: ' + row['reference'])
    if not pd.isna(row['documentation']):
        links.append('documentation: ' + row['documentation'])
    if not pd.isna(row['wikipedia']):
        links.append('wikipedia: ' + row['wikipedia'])

  # Join the links list with a separator (e.g., comma)
    links_str = ', '.join(links) if links else None

  # Set the value for the 'Links' column
    row['Links'] = links_str
    return row
```

```python
# Get the row at the specified index
df = df.apply(append_links,axis = 1)
```
